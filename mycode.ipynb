```
from pyspark.sql import SparkSession
from pyspark.sql.functions import year, month, dayofmonth, dayofweek, to_date, col, expr, when, avg, count
```

```
# Create a Spark session
spark = SparkSession.builder.appName("TransformDatasets").getOrCreate()
```

```
# List of DataFrames
dataframes = [spark_df1, spark_df2, spark_df3, spark_df4, spark_df5, spark_df6]
```

```
# Transformation loop
for i, df in enumerate(dataframes, start=1):
    print(f"Transforming DataFrame spark_df{i}")

    # Convert "Date" column to date format
    df = df.withColumn("Date", to_date("Date", "yyyy-MM-dd"))

    # Extract year, month, day, and day of week
    df = df.withColumn("Year", year("Date"))
    df = df.withColumn("Month", month("Date"))
    df = df.withColumn("Day", dayofmonth("Date"))
    df = df.withColumn("DayOfWeek", dayofweek("Date"))

    # Filter out invalid ArrivalTime and DepartureTime
    df = df.filter(~(col("ArrivalTime").substr(1, 2) >= "24"))
    df = df.filter(~(col("DepartureTime").substr(1, 2) >= "24"))

    # Calculate duration
    df = df.withColumn("Duration", expr(
        "from_unixtime(unix_timestamp(ArrivalTime, 'HH:mm') - unix_timestamp(DepartureTime, 'HH:mm'), 'HH:mm')"
    ))

    # Categorize delays
    df = df.withColumn("DelayCategory", 
                       when(col("Delay") <= 0, "No Delay")
                       .when((col("Delay") > 0) & (col("Delay") <= 10), "Short Delay")
                       .when((col("Delay") > 10) & (col("Delay") <= 20), "Medium Delay")
                       .otherwise("Long Delay"))

    # Calculate average passengers
    average_passengers = df.select(avg("Passengers")).first()[0]

    # Identify peak and off-peak hours
    df = df.withColumn("HeureDePointe", when(col("Passengers") > average_passengers, "peak").otherwise("off-peak"))

    # Calculate average delay, average passengers, and total trips by route
    result_df = df.groupBy("Route").agg(
        avg("Delay").alias("RetardMoyen"),
        avg("Passengers").alias("NombrePassagersMoyen"),
        count("*").alias("NombreTotalVoyages")
    )

    # Display the resulting DataFrame
    display(result_df)
```
